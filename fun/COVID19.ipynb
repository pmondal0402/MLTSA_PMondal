# Polynomial fit

```
date_DE = np.array(data_DE_odr['date'])
pos_DE = np.log(np.array(data_DE_odr['positive']))
# Note first 5 days had no cases and hence log of that data gives inf 
# i.e. log(0)
rpos_DE = pos_DE[6:]
ind = np.arange(len(rpos_DE)) # Index 
plot(ind, rpos_DE, 'bo')
```

```
# Do a linear fit using polynomial fit

def line(x, slope, intercept):
    """ creates a line 
    x: exogenous variable
    slope: slope line parameter
    intercept: intercept line parameter
    """
    return x * slope + intercept

pars = np.polyfit(ind, rpos_DE, 1)
# print(pars)

pl.plot(ind, rpos_DE, 'bo', label='data')
pl.plot(ind, line(ind, pars[0], pars[1]), 'k-', label="polyfit")

# Fig property
pl.legend();
pl.rcParams["figure.figsize"] = (18,10)
pl.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
pl.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
pl.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
pl.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
pl.title("Polynomial fit of order 1 (linear)")
axis_font = {'fontname':'Computer Modern Roman', 'size':'24'}
pl.xlabel("Days", **axis_font)
pl.ylabel("Positive Cases", **axis_font)
pl.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
pl.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```
# Predict future
len_pred = 2 # How much in future you want to predict
             # 2 --> Double the time of dataset

pl.plot(ind, rpos_DE, 'bo', label="Data")
pl.plot(ind, line(ind, pars[0], pars[1]), 'k-', label="polyfit")

# predict
ind_prd = np.arange(len_pred*len(ind))
pl.plot(ind_prd, line(ind_prd, pars[0], pars[1]), 'r-', label="Predict")

# Fig property
pl.legend();
pl.rcParams["figure.figsize"] = (18,10)
pl.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
pl.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
pl.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
pl.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
pl.title("Polynomial fit of order 1 (linear)")
axis_font = {'fontname':'Computer Modern Roman', 'size':'24'}
pl.xlabel("Days", **axis_font)
pl.ylabel("Positive Cases", **axis_font)
pl.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
pl.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```
